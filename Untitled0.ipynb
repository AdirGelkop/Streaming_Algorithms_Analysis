Created a data stream of 100k numbers.
The stream behaves like real-world instances (zipf) - few instances appear most of the time, most instances appear only once or twice.

F2 - sum of squared freq.
Freq - a dict telling how many times each instance appears.

```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

# Data Stream Generation - Zipf Distribution
# Zipf param for total of uniques
# Seed just consistency

def generate_zipf_stream(n_elements=100_000, zipf_param=1.1, seed=42):
    np.random.seed(seed)
    stream = np.random.zipf(zipf_param, n_elements)
    n_unique = len(np.unique(stream))
    return stream, n_unique

# Ground Truth Functions
# Compute the second FM - F2

def compute_f2(stream):
    freq = Counter(stream)
    f2 = sum(f ** 2 for f in freq.values())
    return f2

def compute_frequencies(stream):
    return dict(Counter(stream))

# ---

# Generate the stream
stream, n_unique = generate_zipf_stream(n_elements=100_000, zipf_param=1.1)

# Compute ground truth values
f2_true = compute_f2(stream)
frequencies = compute_frequencies(stream)

# Verify the output
print("=" * 50)
print("STREAM STATISTICS")
print("=" * 50)
print(f"Stream length:      {len(stream):,}")
print(f"Unique elements:    {n_unique:,}")
print(f"F2 (true):          {f2_true:,}")
print(f"Top 5 frequencies:  {sorted(frequencies.values(), reverse=True)[:5]}")
print("=" * 50)

#print("Requirements check:")
#print(f"  Stream >= 100,000: {'✓' if len(stream) >= 100_000 else '✗'}")
#print(f"  Unique >= 10,000:  {'✓' if n_unique >= 10_000 else '✗'}")
#print("=" * 50)
```

AMS Algo to estimate F2 (sum of squared freq) without storing all the data.
We assign each unique element a random value of +1 or -1. Multiply each element's frequency by its hash value, sum everything up, and square the result.
Mathematically, this gives an unbiased estimate of F2.

3 Versions :
1. Alpha - Do it once. Simple but very unstable (high variance).
2. Beta - Do it k times and take the average. More stable.
3. Final - Do Beta m times and take the median. Very stable.

```python
# AMS - Three versions (Alpha, Beta, Final)
# Estimating F2

class AMSEstimators:
    def __init__(self, unique_elements, frequencies):
        self.unique_elements = unique_elements
        self.frequencies = frequencies
        # Create frequency vector aligned with unique elements
        self.freq_vector = np.array([frequencies[e] for e in unique_elements])

    # Generate random {-1, +1} hash values for each unique element
    # n_hashes will be the number of independent hash funcs
    # We return hash_matrix, which is shape [n_unique, n_hashes] of values in {-1, +1}
    def _generate_hash_values(self, n_hashes):
        return np.random.choice([-1, 1], size=(len(self.unique_elements), n_hashes))

    # Alpha - Single basic estimator X^2
    # X = sum(f_i * h(i)) for all elements i
    def alpha(self, n_runs=50):
        # Generate one hash per run
        H = self._generate_hash_values(n_runs)
        X = self.freq_vector @ H
        estimates = X ** 2
        return estimates

    # Beta - Average of k independent estimators
    # Reduces variance by factor of k
    def beta(self, k=10, n_runs=50):
        # Generate k hashes per run
        H = self._generate_hash_values(k * n_runs)
        X = self.freq_vector @ H  # shape: (k * n_runs,)
        X_squared = (X ** 2).reshape(n_runs, k)
        estimates = X_squared.mean(axis=1)
        return estimates

    # Final - Median of m groups, each averaging k estimators
    # Provides guarantees (epsilon, delta)
    def final(self, k=10, m=11, n_runs=50):
        # Generate k*m hashes per run
        H = self._generate_hash_values(k * m * n_runs)
        X = self.freq_vector @ H  # shape: (k * m * n_runs,)
        X_squared = (X ** 2).reshape(n_runs, m, k)
        group_means = X_squared.mean(axis=2)  # shape: (n_runs, m)
        estimates = np.median(group_means, axis=1)  # shape: (n_runs,)
        return estimates

# ---

# Initialize estimator with our stream data
unique_elements = np.array(list(frequencies.keys()))
ams = AMSEstimators(unique_elements, frequencies)

# Run each version
alpha_estimates = ams.alpha(n_runs=50)
beta_estimates = ams.beta(k=10, n_runs=50)
final_estimates = ams.final(k=10, m=11, n_runs=50)

# Display results
print("=" * 60)
print("AMS ESTIMATOR VERIFICATION")
print("=" * 60)
print(f"True F2: {f2_true:,}")
print("-" * 60)
print(f"{'Version':<10} {'Mean':>15} {'Std':>15} {'Error %':>12}")
print("-" * 60)

for name, est in [("Alpha", alpha_estimates),
                   ("Beta", beta_estimates),
                   ("Final", final_estimates)]:
    mean_est = est.mean()
    std_est = est.std()
    error_pct = abs(mean_est - f2_true) / f2_true * 100
    print(f"{name:<10} {mean_est:>15,.0f} {std_est:>15,.0f} {error_pct:>11.2f}%")

print("=" * 60)
```

סיכום שלב 3 - מימוש Count-Min Sketch:
מימשנו את אלגוריתם Count-Min Sketch להערכת שכיחויות אלמנטים בזרם. המבנה הוא מטריצת מונים בגודל d×w (עומק × רוחב), כאשר לכל שורה יש פונקציית hash אידיאלית שממפה אלמנטים לעמודות. במקום hash אמיתי, הגרלנו מיפוי קבוע מראש לכל אלמנט ייחודי. בעדכון, כל שורה מגדילה את המונה בעמודה המתאימה, ובשליפה לוקחים את המינימום מכל השורות כדי לצמצם over-estimation מהתנגשויות. הקוד מותאם לביצועים עם np.bincount לעדכון וקטורי מהיר. האימות מראה את התכונה המרכזית של האלגוריתם: השגיאה תמיד חיובית (one-sided error) - האומדן תמיד גדול או שווה לשכיחות האמיתית, מה שהופך אותו לשימושי במיוחד כשחשוב לא לפספס (למשל זיהוי heavy hitters).

```python
# =============================================================================
# Cell 3: Count-Min Sketch Implementation
# =============================================================================

class CountMinSketch:
    """
    Count-Min Sketch for frequency estimation.
    Returns minimum count across d hash functions to reduce over-estimation.
    """

    def __init__(self, w, d, unique_elements):
        """
        Initialize Count-Min Sketch.

        Args:
            w: width (number of columns/buckets per row)
            d: depth (number of rows/hash functions)
            unique_elements: array of unique element IDs for hash mapping
        """
        self.w = w
        self.d = d
        self.unique_elements = unique_elements
        self.n_unique = len(unique_elements)

        # Create element to index mapping for fast lookup
        self.elem_to_idx = {e: i for i, e in enumerate(unique_elements)}

        # Pre-generate ideal hash functions: random mapping for each element
        # hash_matrix[i, j] = column index for element j in row i
        self.hash_matrix = np.random.randint(0, w, size=(d, self.n_unique))

        # Initialize counter matrix
        self.counters = np.zeros((d, w), dtype=np.int64)

    def update_all(self, frequencies):
        """
        Update all counters at once using vectorized operations.

        Args:
            frequencies: dict mapping element -> frequency
        """
        # Create frequency vector aligned with unique elements
        freq_vector = np.array([frequencies[e] for e in self.unique_elements])

        # For each row, add frequencies to appropriate columns
        for row in range(self.d):
            # Use np.bincount to sum frequencies by column index
            col_indices = self.hash_matrix[row]
            row_counts = np.bincount(col_indices, weights=freq_vector, minlength=self.w)
            self.counters[row] += row_counts.astype(np.int64)

    def query(self, element):
        """
        Query frequency estimate for a single element.
        Returns minimum count across all rows.

        Args:
            element: element ID to query

        Returns:
            estimated frequency (always >= true frequency)
        """
        idx = self.elem_to_idx[element]
        counts = [self.counters[row, self.hash_matrix[row, idx]] for row in range(self.d)]
        return min(counts)

    def query_batch(self, elements):
        """
        Query frequency estimates for multiple elements (vectorized).

        Args:
            elements: array of element IDs

        Returns:
            array of estimated frequencies
        """
        indices = np.array([self.elem_to_idx[e] for e in elements])

        # Gather counts for all elements across all rows
        # Shape: (d, n_elements)
        all_counts = np.zeros((self.d, len(elements)), dtype=np.int64)

        for row in range(self.d):
            col_indices = self.hash_matrix[row, indices]
            all_counts[row] = self.counters[row, col_indices]

        # Return minimum across rows
        return all_counts.min(axis=0)


# -----------------------------------------------------------------------------
# Verify Count-Min Sketch Implementation
# -----------------------------------------------------------------------------

# Test with moderate sketch size
cms = CountMinSketch(w=1000, d=5, unique_elements=unique_elements)
cms.update_all(frequencies)

# Query some elements and compare to true frequencies
test_elements = list(frequencies.keys())[:10]
estimates = cms.query_batch(np.array(test_elements))
true_freqs = [frequencies[e] for e in test_elements]

print("=" * 60)
print("COUNT-MIN SKETCH VERIFICATION")
print("=" * 60)
print(f"Sketch size: w={cms.w}, d={cms.d}")
print(f"Total counters: {cms.w * cms.d:,}")
print("-" * 60)
print(f"{'Element':<12} {'True Freq':>12} {'Estimate':>12} {'Error':>12}")
print("-" * 60)

for elem, true_f, est in zip(test_elements, true_freqs, estimates):
    error = est - true_f  # Always >= 0 for Count-Min
    print(f"{elem:<12} {true_f:>12,} {est:>12,} {error:>+12,}")

print("-" * 60)
print("Note: Count-Min always over-estimates (error >= 0)")
print("=" * 60)
```

```python
# =============================================================================
# Cell 4.1: Simulation Execution (Memory-Efficient)
# =============================================================================

import gc  # Garbage collector for memory management

N_RUNS = 50

# AMS parameters
AMS_K_VALUES = [1, 5, 10, 20, 50]
AMS_M = 7  # Number of groups for median

# Count-Min parameters
CMS_W_VALUES = [100, 500, 1000, 5000, 10000]
CMS_D = 5

# -----------------------------------------------------------------------------
# Memory-Efficient AMS Experiment
# -----------------------------------------------------------------------------

def run_single_ams_trial(freq_vector, n_unique, k, m):
    """
    Run a single AMS trial using loop-based approach to save memory.
    Returns a single F2 estimate.
    """
    group_means = []

    for g in range(m):
        # Compute k estimators for this group
        estimators = []
        for _ in range(k):
            # Generate hash for this single estimator
            h = np.random.choice([-1, 1], size=n_unique)
            X = np.dot(freq_vector, h)
            estimators.append(X ** 2)

        # Mean of k estimators in this group
        group_means.append(np.mean(estimators))

    # Median of group means
    return np.median(group_means)

print("=" * 60)
print("RUNNING AMS EXPERIMENTS")
print("=" * 60)

ams_results = {'k': [], 'bias': [], 'norm_var': [], 'rel_error': []}
freq_vector = np.array([frequencies[e] for e in unique_elements])
n_unique = len(unique_elements)

for k in AMS_K_VALUES:
    estimates = []

    for run in range(N_RUNS):
        est = run_single_ams_trial(freq_vector, n_unique, k, AMS_M)
        estimates.append(est)

    estimates = np.array(estimates)

    # Compute metrics
    bias = estimates.mean() - f2_true
    norm_var = estimates.var() / f2_true
    rel_error = np.abs(estimates - f2_true).mean() / f2_true

    ams_results['k'].append(k)
    ams_results['bias'].append(bias)
    ams_results['norm_var'].append(norm_var)
    ams_results['rel_error'].append(rel_error)

    print(f"k={k:3d}: Bias={bias/f2_true:+.4f}, NormVar={norm_var:.2e}, RelErr={rel_error:.4f}")
    gc.collect()  # Free memory

print("\n" + "=" * 60)
print("RUNNING COUNT-MIN EXPERIMENTS")
print("=" * 60)

# -----------------------------------------------------------------------------
# Memory-Efficient Count-Min Experiment
# -----------------------------------------------------------------------------

cms_results = {'w': [], 'bias': [], 'norm_var': [], 'rel_error': []}

# Sample elements: top 50 + bottom 50 by frequency
sorted_elems = sorted(frequencies.keys(), key=lambda x: frequencies[x], reverse=True)
sample_elements = np.array(sorted_elems[:50] + sorted_elems[-50:])
true_freqs = np.array([frequencies[e] for e in sample_elements])

for w in CMS_W_VALUES:
    all_errors = []
    all_estimates = []

    for run in range(N_RUNS):
        # Create fresh sketch
        cms = CountMinSketch(w=w, d=CMS_D, unique_elements=unique_elements)
        cms.update_all(frequencies)

        # Query sample elements
        estimates = cms.query_batch(sample_elements)
        all_estimates.append(estimates.copy())
        all_errors.append(estimates - true_freqs)

    all_estimates = np.array(all_estimates)  # Shape: (N_RUNS, n_samples)
    all_errors = np.array(all_errors)

    # Compute metrics (averaged across samples and runs)
    bias = all_errors.mean()
    norm_var = all_estimates.var() / true_freqs.mean()
    rel_error = (np.abs(all_errors) / (true_freqs + 1)).mean()

    cms_results['w'].append(w)
    cms_results['bias'].append(bias)
    cms_results['norm_var'].append(norm_var)
    cms_results['rel_error'].append(rel_error)

    print(f"w={w:5d}: Bias={bias:+.2f}, NormVar={norm_var:.2e}, RelErr={rel_error:.4f}")
    gc.collect()  # Free memory

print("\n" + "=" * 60)
print("EXPERIMENTS COMPLETE!")
print("=" * 60)
```

### Metrics Definition

**Normalized Variance** is defined as specified in the assignment:

$$\text{Normalized Variance} = \frac{Var(\hat{n})}{n}$$

Where:
- For **AMS**: $n = F_2$ (the true second frequency moment = 132,554,912)
- For **Count-Min**: $n = \bar{f}$ (mean true frequency of sampled elements)

Note: The high Normalized Variance values for AMS (e.g., $2.91 \times 10^7$ for k=1) are expected because we divide by $F_2$ which is ~132 million. The key observation is the **relative decrease** in variance as k increases, not the absolute values.

## Summary and Conclusions

### AMS Algorithm: Version Comparison

The three versions of AMS demonstrate a clear progression in estimation quality. **Alpha** (single X² estimator) is unbiased but exhibits extremely high variance, making individual estimates unreliable. **Beta** (mean of k estimators) maintains unbiasedness while reducing variance by a factor of k through averaging—our experiments showed variance dropping from ~29M (k=1) to ~580K (k=50). **Final** (median of means) adds robustness against outliers: by taking the median of m independent group averages, extreme values from any single group cannot skew the result. This "median trick" transforms a high-probability bound into a guaranteed (ε,δ) bound, where δ decreases exponentially with m.

### AMS vs. Count-Min Sketch: Fundamental Differences

The key distinction lies in their error characteristics. **AMS is unbiased**: the {-1, +1} hash values ensure that E[X²] = F₂ exactly, as cross-terms cancel out in expectation. Our normalized bias remained within ±3% of zero across all configurations. **Count-Min has one-sided error**: hash collisions can only add to counters, never subtract, resulting in systematic over-estimation. This positive bias decreases with larger width (fewer collisions) but never reaches zero. The choice between them depends on the use case: AMS for accurate global statistics, Count-Min when over-estimation is acceptable (e.g., conservative frequency thresholds).

### Empirical vs. Theoretical Bounds

Our experiments consistently showed empirical errors well below theoretical worst-case guarantees. For AMS with k=50, the empirical relative error was 0.058 versus the theoretical bound of 0.283. For Count-Min with w=10,000, empirical relative error was 0.56 versus a much higher theoretical prediction. This gap exists because: (1) theoretical bounds assume adversarial inputs while Zipf-distributed data is "easier," and (2) bounds are designed for worst-case guarantees with high probability, not average-case performance. The theory remains valuable as it provides reliable guarantees regardless of input distribution.

### Memory-Efficiency Note

During development, we encountered memory limitations in Google Colab when using fully vectorized numpy operations. The solution was switching to a loop-based approach that processes one trial at a time, trading some speed for stability. This highlights a practical consideration in streaming algorithms: while theoretical analysis focuses on sketch size, real implementations must also consider computational overhead.
